```python
origin_dict = {}
for ori, ori2 in zip(a['originals'], a['originals2']):
    origin_dict[ori] = ori2
    
print(origin_dict)
```

```python
import os
all_filenames = os.listdir(r'C:\Users\28056\Desktop\研究生\AnimeTAB\AnimeTAB\AnimeTAB\Entire songs')
all_filenames = [i for i in all_filenames if '.xml' in i]
for i, j in zip(all_filenames, songs):
    print(i, j.info.original)
    
    
```

```python
for songnum, song in enumerate(songs):
    old_original = song.info.original
    try:
        new_original = origin_dict[old_original]
    except:
        new_original = old_original
        
    old_filename = all_filenames[songnum]
    new_filename = old_filename.replace(old_original, new_original)
    old_filenamepath = r'C:\Users\28056\Desktop\研究生\AnimeTAB\AnimeTAB\AnimeTAB\Entire songs\{}'.format(old_filename)
    new_filenamepath = r'C:\Users\28056\Desktop\研究生\AnimeTAB\AnimeTAB\AnimeTAB\Entire songs\{}'.format(new_filename)
    print(old_filenamepath)
    try:
        os.rename(old_filenamepath, new_filenamepath)
    except:
        0
    
```
