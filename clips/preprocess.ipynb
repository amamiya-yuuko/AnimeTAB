```python
#降八度
def octave_change(xmlfile):
    test_tree = parse(xmlfile)
    test_element = test_tree.documentElement
    staff_tuning = test_element.getElementsByTagName('staff-tuning')
    tuning_octaves = []
    for Node in staff_tuning:
        tuning_octaves.append(Node.getElementsByTagName('tuning-octave')[0].firstChild.data)

    if tuning_octaves == ['5', '4', '4', '4', '3', '3']:
        print('{} has high octave'.format(xmlfile))
        for Node in staff_tuning:
            octave = Node.getElementsByTagName('tuning-octave')[0].firstChild.data
            Node.getElementsByTagName('tuning-octave')[0].firstChild.data = int(octave) - 1
            
        with open(xmlfile, 'w', encoding='utf-8') as f:
            test_tree.writexml(f, addindent=' ', newl='\n', encoding='utf-8')
        
for filename in full_xml_filenames:
    octave_change(path+filename)
```
