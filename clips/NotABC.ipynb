```python
#将没有分ABC的片段单跳出来
import shutil
import os

path = os.getcwd()
newpath = path + '\\havnt'
filenames = [i for i in os.listdir(path) if '.xml' in i]
```

```python
for file in filenames:
    if file.split('.')[0][-1] not in ['A', 'B', 'C', 'I']:
        print(file)
```

```python
for file in filenames:
    if file.split('.')[0][-1] not in ['A', 'B', 'C', 'I']:
        shutil.move(path + '\\' + file, newpath+'\\'+file)
```
